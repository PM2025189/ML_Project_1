```python
id_cols = ["country", "country_code", "year"]

# last look at the cleaned dataframe
print("Columns:", list(df_wdi_clean.columns))   
print("Shape:", df_wdi_clean.shape)
print("\nTypes :")
print(df_wdi_clean.dtypes)

print("\nNA par colonne (desc) :")
print(df_wdi_clean.isna().sum().sort_values(ascending=False))

print("\nColonnes non numériques (hors id) :")
non_num = (df_wdi_clean.columns.difference(id_cols)
           .difference(df_wdi_clean.select_dtypes(include=["number","bool"]).columns))
print(list(non_num))
```

```python
# convert features to numeric, coercing errors to NaN
feat = df_wdi_clean.columns.difference(id_cols)

df_wdi_clean[feat] = (df_wdi_clean[feat]
                      .replace({r",": ""}, regex=True)  # supprime séparateurs de milliers
                      .apply(pd.to_numeric, errors="coerce"))

# Colonnes qui ont généré des NaN après conversion (suspects)
suspects = df_wdi_clean[feat].isna().sum()
print(suspects[suspects > 0].sort_values(ascending=False))
```

```python
# checking that all features (except IDs) are numeric
assert df_wdi_clean[feat].select_dtypes(exclude=["number","bool"]).empty, \
       f"Non numériques: {list(df_wdi_clean[feat].select_dtypes(exclude=['number','bool']).columns)}"
```

```python
# make sure ID columns are of appropriate types

id_cols = [c for c in ["country", "country_code", "year"] if c in df_wdi_clean.columns]
if "year" in df_wdi_clean.columns:
    df_wdi_clean["year"] = pd.to_numeric(df_wdi_clean["year"], errors="coerce").astype("Int64")
```

```python
# Both columns are guaranteed to exist and be numeric.(che = current health expenditure)
che_pct = "Current health expenditure (% of GDP)"
che_pc  = "Current health expenditure per capita (current US$)"


# convert % -> share in [0,1]
df_wdi_clean["health_exp_share"] = df_wdi_clean[che_pct] / 100.0

# stabilize heavy-tailed money values (safe: no log(0) since che_pc > 0)
df_wdi_clean["log1p_health_exp_pc"] = np.log1p(df_wdi_clean[che_pc])

# GDP_pc = 100 * (che_pc / (che_pct / 100)) = 100 * che_pc / che_pct
# only if che_pct > 0 and che_pc is not NaN
mask = df_wdi_clean[che_pct].gt(0) & df_wdi_clean[che_pc].notna()
df_wdi_clean["GDP_pc_imputed"] = np.where(
    mask,
    100 * df_wdi_clean[che_pc] / df_wdi_clean[che_pct],
    np.nan
)
```

```python

# Ensure year numeric for sorting/splits

df_wdi_clean["year"] = pd.to_numeric(df_wdi_clean["year"], errors="coerce").astype("Int64")

target_col = "Mortality rate, infant (per 1,000 live births)"
assert target_col in df_wdi_clean.columns, "Target missing."

df_wdi_clean = df_wdi_clean.sort_values(["country_code","year"])
df_wdi_clean["y_t1"] = df_wdi_clean.groupby("country_code")[target_col].shift(-1)
```

```python
# AR(1) of the target
df_wdi_clean[f"{target_col}_lag1"] = (
    df_wdi_clean.groupby("country_code")[target_col].shift(1)
)

# Candidate numeric predictors at time t (exclude year, target, label)
base_feats = df_wdi_clean.select_dtypes(include=["number"]).columns.difference(
    ["year", target_col, "y_t1"]
)

# Lag-1 for each numeric predictor
for c in base_feats:
    df_wdi_clean[f"{c}_lag1"] = df_wdi_clean.groupby("country_code")[c].shift(1)

# YoY % change for ALL predictors at once, with NO implicit forward-fill
# (keeps NaNs where data is missing; safer for forecasting)
yoy = (df_wdi_clean
       .groupby("country_code", observed=True)[base_feats]
       .pct_change(fill_method=None))

# Clean +/-inf produced by division by zero
yoy = yoy.replace([np.inf, -np.inf], np.nan)

# Suffix columns and attach to df
yoy.columns = [f"{c}_yoy" for c in yoy.columns]
df_wdi_clean = pd.concat([df_wdi_clean, yoy], axis=1)
```

```python
# Final shapes
id_cols = [c for c in ["country","country_code","year"] if c in df_wdi_clean.columns]
mask = df_wdi_clean["y_t1"].notna()

X_cols = df_wdi_clean.columns.difference(id_cols + [target_col, "y_t1"])
X = df_wdi_clean.loc[mask, X_cols]
y = df_wdi_clean.loc[mask, "y_t1"]

print("Panel shapes -> X:", X.shape, "| y:", y.shape)
```
